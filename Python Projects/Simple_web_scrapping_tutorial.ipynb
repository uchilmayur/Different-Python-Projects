```python
import requests # pip install requests
from bs4 import BeautifulSoup as bs # pip install beautifulsoup4
```

```python
r = requests.get("https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=data+analyst&txtLocation=")

# Convert to a beautiful soup object
soup = bs(r.content, 'lxml')

# Print out our html
#print(soup.prettify())
```

```python
# we first  did job = soup.find then carried on with soup.find_all & jobs 
# li clearfix job-bx wht-shd-bx
jobs = soup.find_all('li' , class_ = 'clearfix job-bx wht-shd-bx')
#print(job)
for job in jobs :

  #joblist-comp-name h3
  company_name = job.find('h3',class_ = 'joblist-comp-name').text.strip()
  #print(company_name.text.strip())


  # span srp-skills
  skills = job.find('span' , class_ = 'srp-skills').text.strip()
  #print(skills.text.strip())


  # span sim-posted
  job_posted = job.find('span' , class_ = 'sim-posted').text.strip()



  print(f''' 
  Company name : {company_name}
  Skills : {skills}
  Posted on : {job_posted}
  ''')
  print("*"*200)
```
